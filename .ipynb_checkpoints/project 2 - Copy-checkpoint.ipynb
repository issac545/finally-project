```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
data=pd.read_csv("Fat_Supply_Quantity_Data.csv")
data
```

```python
data .isnull().sum()
```

```python
data
```

```python
data.info()
```

```python
data.Undernourished=data.Undernourished.str.replace("<","")
data.Undernourished=data.Undernourished.astype('float')

```

```python
data.rename({"Undernourished":"< Undernourished"},axis=1,inplace=True)
```

```python
data
```
