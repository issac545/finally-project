```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("movies.csv")
data
```

```python
data.isnull().sum()
```

```python
# Y = a + bX  train 

x=data.loc[:700,["Votes"]]
y=data.loc[:700,["Rating"]]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x_test=data.loc[701:,["Votes"]]
y_test=data.loc[701:,["Rating"]]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
# error = y_test - y_predict 
```

```python
import sklearn.metrics as mc
```

```python
dir(mc)
```

```python
er =mc.mean_absolute_error(y_test,y_predict)
er
```

```python
plt.scatter(data.Votes,data.Rating)
plt.plot(x_test,y_predict,color="red")
```
