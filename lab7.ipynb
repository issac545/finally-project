```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x=np.arange(0,10)
y=np.sin(x)
```

```python
plt.plot(x,y,color="red",linestyle=":",marker=">",markerfacecolor='blue',markersize=12,linewidth=5)
plt.xlabel(" X ")
plt.ylabel(" Y =  sin ( x ) ")
plt.title(" sin wave ",fontsize=20, color='y',loc='left',rotation=45, backgroundcolor="green")
plt.show()
```

```python
z=np.cos(x)
plt.plot(x,z,'b:o',label=" cos x ") # [color][line][marker]
plt.plot(x,y,'y--<', label=" sin ")
plt.legend(loc="best")
plt.show()
```
