```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn 
```

```python
data=pd.read_csv("movies.csv")
data
```

```python
data.isna().sum()
```

```python
data.interpolate(inplace=True)
```

```python
data.isna().sum()
```

```python
data
```

```python
x=data[["Votes","Metascore"]]  # y=a+bx       y=a +b1x1 +b2x2
y=data.Rating
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
x_test
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data,x_vars=["Votes","Metascore"],y_vars=["Rating"],diag_kind=None,kind="scatter",size=5)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
y_test
```

```python
model.intercept_
```

```python
# statsmodels 

import statsmodels.formula.api as sm
```

```python
lr=sm.ols(formula='Rating ~ Votes + Metascore',data=data[:800]).fit() #  x="Votes","Metascore"  y=Rating  y ~ x
```

```python
y_sm_predict=lr.predict({"Votes":4.555,"Metascore":9.5})
y_sm_predict
```
