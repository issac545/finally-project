```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dt=sns.load_dataset("tips")
```

```python
plt.scatter(dt.total_bill,dt.tip)
plt.show()
```

```python
sns.relplot(data=dt,x="total_bill",y="tip",kind="scatter",hue="smoker") # default = scatter
```

```python
sns.relplot(data=dt,x="total_bill",y="tip",kind="scatter",hue="day",col="smoker") # default = scatter
```
